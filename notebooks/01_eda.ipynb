# Olist Customer Retention Analysis - EDA

## 1. Import Library

```python
# Import library
import pandas as pd
import matplotlib as plt
import seaborn as sns
from dotenv import load_dotenv
import os
from sqlalchemy import create_engine
```

## 2. Load Environment & Database Connection

```python
load_dotenv()

DB_HOST = os.getenv("DB_HOST")
DB_USER = os.getenv("DB_USER")
DB_PASSWORD = os.getenv("DB_PASSWORD")
DB_PORT = os.getenv("DB_PORT")
DB_NAME = os.getenv("DB_NAME")
DB_NAME_ANALYTICS = os.getenv("DB_NAME_ANALYTICS")
```

```python
# koneksi ke MySQL
engine = create_engine(f"mysql+pymysql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}")
```

## 3. Load Dataset

```python
customers = "SELECT * FROM olist_raw.raw_customers"
orders = "SELECT * FROM olist_raw.raw_orders"
order_items = "SELECT * FROM olist_raw.raw_order_items"
order_payments = "SELECT * FROM olist_raw.raw_order_payments"
order_reviews = "SELECT * FROM olist_raw.raw_order_reviews"
products = "SELECT * FROM olist_raw.raw_products"
product_category_translation = "SELECT * FROM olist_raw.raw_product_category_translation"
```

```python
# dataset orders
df_orders = pd.read_sql(orders, engine)
df_orders.head()
```

```python
# dataset customers
df_customers = pd.read_sql(customers, engine)
df_customers.head()
```

```python
# dataset order_items
df_order_items = pd.read_sql(order_items, engine)
df_order_items.head()
```

```python
# dataset order_payments
df_order_payments = pd.read_sql (order_payments, engine)
df_order_payments.head()
```

```python
# dataset order_reviews
df_order_reviews = pd.read_sql(order_reviews, engine)
df_order_reviews.head()
```

```python
# dataset products
df_products = pd.read_sql(products, engine)
df_products.head()
```

```python
# dataset product_category_translation
df_product_category_translation = pd.read_sql(product_category_translation, engine)
df_product_category_translation.head()
```
